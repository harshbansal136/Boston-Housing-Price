# CORRELATION AND PAIRPLOT

```python
corrmat=data.corr()
corrmat
```

```python
fig, ax=plt.subplots(figsize=(20,10))
sns.heatmap(corrmat,annot=True,annot_kws={'size':12},square=True,cbar=True)
```

```python
def getCorrelatedFeature(corrdata,threshold):
    feature=[]
    value=[]
    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index])>threshold:
            feature.append(index)
            value.append(corrdata[index])
    df=pd.DataFrame (data=value,index=feature,columns=['Corr Value'])
    return df
```

```python
threshold=0.60
corr_value= getCorrelatedFeature(corrmat['Price'],threshold)
corr_value
```

```python
correlated_data=data[corr_value.index]
correlated_data.head()
```

```python
g = sns.pairplot(correlated_data)
g.map_upper(sns.scatterplot,color='b')
g.map_lower(sns.scatterplot, color='y')
g.map_diag(plt.hist, color='teal',edgecolor = 'k')
plt.tight_layout
```

```python
rows=1
cols=3
fig,ax=plt.subplots(nrows=rows,ncols=cols,figsize=(8,4))

col=correlated_data.columns
index=0
for i in range(rows):
    for j in range (cols):
        sns.distplot(data[col[index]],ax=ax[j],color='teal')
        index=index+1
plt.tight_layout()
```

# STANDARDISATION AND DATA SPLIT

```python
y=correlated_data['Price']
X=correlated_data.drop(columns=['Price'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=0)
X_train.shape,X_test.shape
```

# MODELS

```python
# Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
LinReg=LinearRegression()
LinReg.fit(X_train,y_train)
y_predict=LinReg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
score_lr=r2_score(y_test,y_predict)
mae=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
print('r2: ',score_lr)
print('mae: ',mae)
print('mse: ',mse)
```

```python
# Decision Tree
```

```python
from sklearn.tree import DecisionTreeRegressor
DeTree= DecisionTreeRegressor(min_samples_leaf=15,max_depth=5,random_state = 1001)
DeTree.fit(X_train,y_train)
y_predict=DeTree.predict(X_test)
```

```python
score_dt=r2_score(y_test,y_predict)
mae=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
print('r2: ',score_dt)
print('mae: ',mae)
print('mse: ',mse)
```

```python
# Random Forests
```

```python
from sklearn.ensemble import RandomForestRegressor
Forest= RandomForestRegressor(bootstrap= [True],n_estimators=20,random_state=115)
Forest.fit(X_train,y_train)
y_predict=Forest.predict(X_test)
score_rf=r2_score(y_test,y_predict)
mae=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
print('r2: ',score_rf)
print('mae: ',mae)
print('mse: ',mse)
```

```python
# KNN Algorithm
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
error_rate = []
for i in range(1,60):
    
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
 
plt.figure(figsize=(10,6))
plt.plot(range(1,60),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
KNN = KNeighborsRegressor(n_neighbors=8,p=2)
KNN.fit(X_train,y_train)
pred_y = KNN.predict(X_test)
score_knn=r2_score(y_test,pred_y)
mae=mean_absolute_error(y_test,pred_y)
mse=mean_squared_error(y_test,pred_y)
print('r2: ',score_knn)
print('mae: ',mae)
print('mse: ',mse)
```

```python
# SVM Algorithm 
```

```python
from sklearn.svm import SVR
Svr= SVR(kernel='rbf',gamma='auto')
Svr.fit(X_train,y_train)
y_predict=Svr.predict(X_test)
score_svm=r2_score(y_test,y_predict)
mae=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
print('r2: ',score_svm)
print('mae: ',mae)
print('mse: ',mse)
```

```python
models = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Random Forests', 
              'KNN', 'SVM'],
    'Score': [score_lr,score_dt,score_rf,score_knn,score_svm]})
models=models.sort_values(by='Score', ascending=False).reset_index()
model=models.drop(columns=['index'])
model
```
